```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# pretty print
from pprint import pprint

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count, store in a variable to use in a loop
cities_returned = len(cities)
cities_returned
```

```python
# look at cities generated
cities
```

```python
# 5-25 this gets the data from the api

# get the base url
base_url = "http://api.openweathermap.org/data/2.5/weather?appid="

# declare open arrays
lat = []
temp = []
humidity = []
cloudiness = []
wind_speed = []

# get the data
for city in cities:
    weather_data = requests.get(base_url + api_key + "&q=" + city + "&units=imperial").json()
    
    try:
        lat.append(weather_data['coord']['lat'])
        # print city message for log
        print(f"Gathering and adding data for city: {city}")
    except KeyError:
        # if lat data missing then null
        lat.append(None)
        # print failed message for log
        print(f"No latitude data for city: {city}--Skipped!")
              
    try:
        temp.append(weather_data['main']['temp'])
    except KeyError:
        # if temp data missing then null
        temp.append(None)
        
    try:
        humidity.append(weather_data['main']['humidity'])
    except KeyError:
        # if humidity data missing then null
        humidity.append(None)
    
    try:
        cloudiness.append(weather_data['clouds']['all'])
    except KeyError:
        # if cloudiness data is missing then null
        cloudiness.append(None)
    
    try:
        wind_speed.append(weather_data['wind']['speed'])
    except KeyError:
        # if wind speed data is missing then null
        wind_speed.append(None)
        
    # need to print the log here of each city as it is processed
    
    # need to save as CSV of what was selected

#print("all done!")    # test for end of code completion
```

```python
# 5-25, develop dict from result above
weather_dict = {
    "city": cities,
    "lat": lat,
    "temp": temp,
    "humidity": humidity,
    "cloudiness": cloudiness,
    "wind speed": wind_speed
}

# convert dictionary to data frame
weather_data = pd.DataFrame(weather_dict)
weather_data.head()
```
